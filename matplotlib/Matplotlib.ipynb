<h1 style="color:orange"><i>Axes Labels, Legend, Grid</i></h1>

```python
# Show legend
plt.plot(days, max_t, label="max")
plt.plot(days, min_t, label="min")
plt.plot(days, avg_t, label="average")

plt.legend(loc='best')
```

```python
# Legend at different location with shadow enabled and fontsize set to large
plt.plot(days, max_t, label="max")
plt.plot(days, min_t, label="min")
plt.plot(days, avg_t, label="average")

plt.legend(loc='upper right',shadow=True,fontsize='large')
```

<h1 style = "color:orange"><i>Bar Chart</i></h1>

```python
import numpy as np
```

## <i>Simple bar chart showing revenues of major US tech companies</i>

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
```

```python
xpos = np.arange(len(company))
xpos
```

```python
plt.bar(xpos,revenue, label="Revenue")

plt.xticks(xpos,company)
plt.ylabel("Revenue(Bln)")
plt.title('US Technology Stocks')
plt.legend()
```

## <i>Multiple Bars showing revenue and profit of major US tech companies</i>

```python
profit=[40,2,34,12]
```

```python
plt.bar(xpos-0.2,revenue, width=0.4, label="Revenue")
plt.bar(xpos+0.2,profit, width=0.4,label="Profit")

plt.xticks(xpos,company)
plt.ylabel("Revenue(Bln)")
plt.title('US Technology Stocks')
plt.legend()
```

### <i>Horizontal bar chart using barh function</i>

```python
plt.barh(xpos,revenue, label="Revenue")

plt.yticks(xpos,company)
plt.ylabel("Revenue(Bln)")
plt.title('US Technolog Stocks')
plt.legend()
```

<h1 style = "color:orange"><i>Histogram</i></h1>

### <b>In histograms, x axis contains a variable and y axis will be a frequency of that variable</b>

### We have a sample data of blood sugar level of different patients, we will try to plot number of patients by blood range and try to figure out how many patients are normal, pre-diabetic and diabetic

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]

plt.hist(blood_sugar, rwidth=0.8) # by default number of bins is set to 10
```

```python
plt.hist(blood_sugar,rwidth=0.5,bins=4)
```

#### 

    80-100: Normal
    100-125: Pre-diabetic
    80-100: Diabetic

###

```python
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")

plt.hist(blood_sugar, bins=[80,100,125,150], rwidth=0.95, color='g')
```

## <i>Mutiple data samples in a histogram</i>

```python
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")

blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]

plt.hist([blood_sugar_men,blood_sugar_women], bins=[80,100,125,150], rwidth=0.95, color=['green','orange'],label=['men','women'])
plt.legend()
```

### <i>Horizontal orientation</i>

```python
plt.xlabel("Number Of Patients")
plt.ylabel("Sugar Level")
plt.title("Blood Sugar Chart")

plt.hist(blood_sugar, bins=[80,100,125,150], rwidth=0.95, color='red', orientation='horizontal')
```

<h1 style = "color:orange"><i>Pie Charts</i></h1>

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
plt.pie(exp_vals,labels=exp_labels)
plt.show() # to remove xtra writings
```

### <i>Show percentages for every pie. Specify radius to increase chart size</i>

```python
plt.pie(exp_vals,labels=exp_labels, autopct='%1.1f%%',radius=1.5)
plt.show()
```

```python
plt.pie(exp_vals,labels=exp_labels, autopct='%1.1f%%',radius=1.5,explode=[0,0.2,0,0.1,0])
plt.show()
```

```python
plt.pie(exp_vals,labels=exp_labels, autopct='%1.1f%%',radius=1,explode=[0,0,0,0.1,0.2],counterclock=True, startangle=45)
plt.show()
```

<h1 style = "color:orange"><i>Saving chart to a file using savefig</i></h1>

```python
plt.pie(exp_vals,labels=exp_labels,radius=2, autopct='%0.1f%%', explode=[0,0.1,0.1,0,0])
plt.savefig("piechart.jpg", bbox_inches="tight", pad_inches=1, transparent=True)
```

### <i>You can also save in pdf format and to a different location on your computer by specifying absolute path</i>

```python
plt.pie(exp_vals,labels=exp_labels,radius=1, autopct='%0.1f%%', explode=[0,0.1,0.1,0,0])
plt.savefig("piechart.pdf", bbox_inches="tight", pad_inches=2, transparent=True)
```
