```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
var = sns.load_dataset("penguins").head(50)
var
```

```python
# To show histogram
sns.displot(var["bill_length_mm"])
plt.show()
```

```python
sns.displot(var["bill_length_mm"], bins = [35, 37, 39, 41, 43, 45])
plt.show()
```

```python
sns.displot(var["bill_length_mm"], kde = True)
plt.show() # kernel density plot line
```

```python
sns.displot(var["bill_length_mm"], kde = True, rug = True)
plt.show() 
```

```python
sns.displot(var["bill_length_mm"], kde = True, color = "g")
plt.show() # kernel density plot line
```

```python
sns.displot(var["bill_length_mm"], kde = True, log_scale = True)
plt.show() # turn data into log scale
```
