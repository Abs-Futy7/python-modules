# Cat Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

var = sns.load_dataset("tips")
var
```

```python
sns.catplot(x = "size", y = "tip", data = var)
plt.show()
```

```python
sns.catplot(x = "day", y = "size", data = var, hue = "sex", kind = "bar")
plt.show()
```

```python
sns.catplot(x = "day", y = "size", data = var, hue = "sex", kind = "violin")
plt.show()
```

```python
sns.catplot(x = "day", y = "size", data = var, hue = "sex", kind = "box")
plt.show()
```
