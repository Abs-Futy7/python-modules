### Scatter-Plot

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
var = sns.load_dataset("penguins")
var
```

```python
sns.scatterplot(x = "bill_length_mm", y = "bill_depth_mm", data = var)
plt.show()
```

```python
sns.scatterplot(x = "bill_length_mm", y = "bill_depth_mm", data = var, hue = "sex")
plt.show()
```

```python
sns.scatterplot(x = "bill_length_mm", y = "bill_depth_mm", data = var, hue = "sex", style = "sex", size = "sex", sizes = (60, 30))
plt.show()
```

```python
m = {"Male" : "*", "Female" : ">"}
sns.scatterplot(x = "bill_length_mm", y = "bill_depth_mm", data = var, hue = "sex", style = "sex", markers = m)
plt.show()
```
