# Pair Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
var = sns.load_dataset("tips")
var
```

```python
sns.pairplot(var)
plt.show()
```

```python
sns.pairplot(var, hue = "sex")
plt.show()
```

```python
sns.pairplot(var, vars = ["total_bill", "tip"], hue = "sex")
plt.show()
```

```python
sns.pairplot(var, vars = ["total_bill", "tip"], hue = "sex", kind = "reg")
plt.show()
```

```python
sns.pairplot(var, vars = ["total_bill", "tip"], hue = "sex", kind = "kde")
plt.show()
```
