```python
#Strings cannot be heatmap, and have to convert into
# two dimension to do heatmap
data = sns.load_dataset("anagrams")
x = data.drop(columns = "attnr", axis = 1).head(10)
x
```

```python
sns.heatmap(x)
plt.show()
```

```python
sns.heatmap(x, cmap = 'gist_heat')
plt.show()
```

```python
sns.heatmap(x, cmap = 'PuOr')
plt.show()
```

```python
sns.heatmap(x, vmin = 0, vmax = 10, annot = True)
plt.show()
```

```python
var = np.linspace(1, 10, 10).reshape(2,5)
var
```

```python
sns.heatmap(var, vmin = 0, vmax = 10, annot = True)
plt.show()
```

```python
y = {"fontsize" : 20, "color" : "y"}
sns.heatmap(var, vmin = 0, vmax = 10, annot = True, annot_kws = y)
plt.show()
```

```python
y = {"fontsize" : 20, "color" : "y"}
sns.heatmap(var, vmin = 0, vmax = 10, annot = True, annot_kws = y, linewidth = 4, linecolor = "b")
plt.show()
```

```python
sns.heatmap(var, vmin = 0, vmax = 10, annot = True, cbar = False)
plt.show()
```
