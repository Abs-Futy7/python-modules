### Bar-Plot

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
var = sns.load_dataset("penguins")
var
```

```python
sns.barplot(x = var.island, y = var.bill_length_mm)
plt.show()
```

```python
sns.barplot(x = "island", y = "bill_length_mm", data = var)
plt.show()
```

```python
sns.barplot(x = "island", y = "bill_length_mm", data = var, hue = "sex")
plt.show()
```

```python
order1 = ["Dream", "Torgersen", "Bisoe"]
sns.barplot(x = "island", y = "bill_length_mm", data = var, hue = "sex", order = order1)
plt.show()
```

```python
sns.barplot(x = "island", y = "bill_length_mm", data = var, hue = "sex", hue_order = ["Female", "Male"])
plt.show()
```

```python
sns.barplot(x = "island", y = "bill_length_mm", data = var, hue = "sex", hue_order = ["Female", "Male"], errorbar=('ci', 90))
plt.show()
```

```python
# Horizontal parameter can't be strings, have to be number
sns.barplot(x = "bill_depth_mm", y = "bill_length_mm", data = var, hue = "sex", orient = "h")
plt.show()
```

```python
sns.barplot(x = "island", y = "bill_length_mm", data = var, hue = "sex", palette = "icefire")
plt.show()
```

```python
sns.barplot(x = "island", y = "bill_length_mm", data = var, hue = "sex", saturation = 0.3)
plt.show()
```

```python
sns.barplot(x = "island", y = "bill_length_mm", data = var, hue = "sex", errcolor = "r")
plt.show()
```

```python
sns.barplot(x = "island", y = "bill_length_mm", data = var, hue = "sex", errwidth = 12)
plt.show()
```

```python
sns.barplot(x = "island", y = "bill_length_mm", data = var, hue = "sex", errcolor = "r", capsize = 0.3)
plt.show()
```

```python
sns.barplot(x = "island", y = "bill_length_mm", data = var, hue = "sex", dodge = False)
plt.show()
```

```python
sns.set(style = "darkgrid")
sns.barplot(x = "island", y = "bill_length_mm", data = var, hue = "sex")
plt.show()
```
