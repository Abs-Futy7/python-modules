<h1>Line Plot</h1>

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
var = [1,2,3,4,5,6,7]
var1 = [2,3,4,1,5,6,7]

x1 = pd.DataFrame({"var":var, "var1":var1})

sns.lineplot(x="var",y="var1", data = x1) #seaborn method
plt.show() # matplotlib
```

```python
y1 = sns.load_dataset("penguins") # loading datsheets from github
y1
```

```python
sns.lineplot(x="bill_length_mm",y="bill_depth_mm", data = y1)
plt.show()
```

```python
sns.lineplot(x="bill_length_mm",y="bill_depth_mm", data = y1, hue = "sex")
plt.show()
```

```python
sns.lineplot(x="bill_length_mm",y="bill_depth_mm", data = y1, hue = "sex", size = 10)
plt.show()
```

```python
sns.lineplot(x="bill_length_mm",y="bill_depth_mm", data = y1, hue = "sex", style="sex")
plt.show()
```

```python
sns.lineplot(x="bill_length_mm",y="bill_depth_mm", data = y1, hue = "sex", style="sex", palette='Accent')
plt.show() #palette changes the color
```

```python
sns.lineplot(x="bill_length_mm",y="bill_depth_mm", data = y1, hue = "sex", style="sex", palette='Accent', markers=["o", ">"])
plt.show()  # to bring markers at sharp points
```

```python
y1 = sns.load_dataset("penguins").head(10) # To minimize data
y1
```

```python
sns.lineplot(x="bill_length_mm",y="bill_depth_mm", data = y1, hue = "sex", size = 10)
plt.show()
```

```python
# To remove legend
sns.lineplot(x="bill_length_mm",y="bill_depth_mm", data = y1, hue = "sex", size = 50, legend = False)
plt.show()
```

```python
sns.lineplot(x="bill_length_mm",y="bill_depth_mm", data = y1, hue = "sex", size = 50)
plt.grid()
plt.title("Dataset")
plt.show()
```
